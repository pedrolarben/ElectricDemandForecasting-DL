```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
### Remove this line after first execution
results = pd.concat([
            pd.read_csv('../files/results/experimental_results_bs64.csv', delimiter=";"), 
            pd.read_csv('../files/results/experimental_results_bs128.csv', delimiter=";"), 
            pd.read_csv('../files/results/experimental_results_bs256.csv', delimiter=";")
        ], ignore_index=True)
results.to_csv('../files/results/experimental_results.csv', sep=";", index=False)
```

```python
results = pd.read_csv('../files/results/experimental_results.csv', delimiter=";")
results
```

```python
metric_columns = ['mse', 'rmse', 'nrmse', 'mae', 'mpe', 'mape',
       'mdape', 'smape', 'smdape', 'mase', 'rmspe', 'rmsse', 'mre', 'rae',
       'mrae', 'std_ae', 'std_ape']
       #'val_mse', 'val_rmse', 'val_nrmse',
       #'val_mae', 'val_mpe', 'val_mape', 'val_mdape', 'val_smape',
       #'val_smdape', 'val_mase', 'val_rmspe', 'val_rmsse', 'val_mre',
       #'val_rae', 'val_mrae', 'val_std_ae', 'val_std_ape']
```

```python
f, ax = plt.subplots(figsize=(20, 16))
sns.heatmap(results[metric_columns].subtract(results[metric_columns].min()).div(results[metric_columns].max()-results[metric_columns].min()), annot=True, fmt="0.2f", linewidths=.5, cmap='RdYlGn_r',ax=ax)
```

```python
for metric in metric_columns:
    results[metric+'_rank'] = results[metric].rank(ascending=True)
results['mean_rank'] = results[[metric+'_rank' for metric in metric_columns]].mean(axis=1)
```

```python
results.sort_values('mean_rank', ascending=True).head(5)[[metric+'_rank' for metric in metric_columns]+['mean_rank']]
```

```python
f, ax = plt.subplots(figsize=(18, 12))
sns.heatmap(results[[metric+'_rank' for metric in metric_columns]], ax=ax, cmap="RdYlGn_r")

f, ax = plt.subplots(figsize=(18, 6))
sns.boxplot(data=results[[metric+'_rank' for metric in metric_columns]].T,   ax=ax)

f, ax = plt.subplots(figsize=(18, 6))
sns.barplot(data=results.sort_values('mean_rank', ascending=False).tail(10)[[metric+'_rank' for metric in metric_columns]].T,   ax=ax)
```
